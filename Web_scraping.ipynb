```python
import os
import pandas as pd 
import numpy as np
import time
from bs4 import BeautifulSoup
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import  Select
url="https://www.nmc.org.in/information-desk/indian-medical-register/"
driver_path='chromedriver.exe'
browser= Chrome(executable_path=driver_path)
browser.get(url)
time.sleep(5)
# browser.find_element(By.XPATH, '//*[@id="main-content"]/div[2]/div/div/div/ul/li[4]/a').click()
# browser.find_element(By.XPATH,'//*[@id="smc_form"]/div[1]/div[2]/div/button').click()
# state media
browser.find_element(By.XPATH,'//*[@id="main-content"]/div[2]/div/div/div/ul/li[4]/a').click()
time.sleep(2)
element_dropdown = browser.find_element(By.XPATH, '//*[@id="smc_form"]/div[1]/div[2]/div/button')
browser.find_element(By.XPATH, '//*[@id="smc_form"]/div[1]/div[2]/div/ul/li[1]/div/input')
time.sleep(2)
browser.find_element(By.XPATH, '//*[@id="smc_form"]/div[1]/div[2]/div/ul/li[1]/div/input')
element_dropdown.send_keys("West Bengal Medical Council")
element_dropdown.send_keys(Keys.ENTER)
time.sleep(2)
browser.find_element(By.XPATH, '//*[@id="smc_form"]/div[1]/div[2]/div/ul/li[42]/a/label').click()
browser.find_element(By.XPATH, '//*[@id="doctor_smc_details"]').click()
time.sleep(2)
#getting to last page we will run it back
time.sleep(3)
browser.find_element(By.XPATH, '//*[@id="doct_info4_paginate"]/span/a[6]').click()
time.sleep(4)

#going to the page for starting scrapping
for i in range(7):
    browser.find_element(By.XPATH,'//*[@id="doct_info4_previous"]').click()
    time.sleep(10)
    

#fetching all the rows and columns from the table
html = browser.find_element(By.XPATH, '//*[@id="doct_info4"]/tbody').get_attribute('innerHTML')
soup = BeautifulSoup(html, 'html.parser')
columns = []
for row in soup.find_all('tr'):
    columns.append(row.find_all('td'))
#fetching the row number where first record of 2010 is present
i=0
for j in columns:
    i=i+1
    if(j[1].text=='2019'):
        break
#Defining empty lists which will store the doctor's details
names=[]
dob = []
dor = []
yoi = []
rn = []
qy = []
un = []

#Empty dictionary for now...
det = dict({"Names": names,
            "Date of Birth": dob,
            "Year of Info": yoi,
            "Date of Registration": dor,
            "Registration Number": rn,
            "Qualification Year": qy,
            "University Name":un})

# loop for fetching the data from all the rows in the table
for k in range(i,int(browser.find_element(By.XPATH, '//*[@id="doct_info4_length"]/label/select/option[1]').text)+1):
    
    #this clicks the 7th element of each row in the table, which is the 'View' button
    browser.find_element(By.XPATH, '//*[@id="doct_info4"]/tbody/tr['+str(k)+']/td[7]/a').click()
    time.sleep(4)
    
    #fetching the inner table data as an html and parsing it with beautiful soup
    html = browser.find_element(By.ID,'doctorBiodata').get_attribute('innerHTML')
    soup = BeautifulSoup(html, 'html.parser')
    
    #appending all the columns for each row in one row of columns1
    columns1 = []
    for row in soup.find_all('tr'):
        columns1.append(row.find_all('td'))
        #storing the required details in the 'det' dictionary
    det["Names"].append(columns1[0][1].text.strip())
    det["Date of Birth"].append(columns1[2][1].text)
    det["Year of Info"].append(columns1[2][3].text)
    det["Registration Number"].append(columns1[3][1].text)
    det["Date of Registration"].append(columns1[3][3].text)
    det["Qualification Year"].append(columns1[5][3].text)
    det["University Name"].append(columns1[6][1].text)
    
    #closing the pop-up window either by using the cross(x) button or the 'Close' button at the bottom
    try:
        browser.find_element(By.XPATH, '//*[@id="doctorModal"]/div[2]/div/div/div[1]/button/span').click()
    except:
        browser.find_element(By.XPATH, '//*[@id="doctorModal"]/div[2]/div/div/div[3]/button[2]').click()
    time.sleep(1)
    
    #printing the name of the doctor for whom the record was fetched
    print(columns1[0][1].text.strip())

#closing the webdriver as our work is finished
browser.close()

#converting the 'det' dictionary into a dataframe
data = pd.DataFrame(det)


```
